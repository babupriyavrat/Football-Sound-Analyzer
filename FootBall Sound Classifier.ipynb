```python
def FootBallSoundClassifier(path,sound_type):
    #valid values of Sound Type:
    #full - full spectrum
    #background - background
    import fnmatch
    import numpy as np
    import re
    import csv
    from pyAudioAnalysis import audioTrainTest as aT
    isSignificant = 0.8 #try different values.
    Maybe= 0.6 #try different values.

    files = [f for f in os.listdir('.') if os.path.isfile(f)]
    with open('football_sound_classifier_data.csv','w') as outfile: 
        columns=["from","to","Classification","Class","probability"]
        w = csv.DictWriter(outfile,fieldnames=columns)
        w.writeheader()
        for f in files:
            if fnmatch.fnmatch(f, '*S_'+sound_type+'*.wav'):
            # P: list of probabilities
                Result, P, classNames = aT.fileClassification(f, "svmModel", "svm")
                winner = np.argmax(P) #pick the result with the highest probability value.

                # is the highest value found above the isSignificant threshhold? 
                if P[winner] > isSignificant :
                    outfile.write(re.split("to",f[16:])[0]+","+re.split("sec",re.split("to",f[16:])[1])[0]+","+"Significant"+","+classNames[winner]+","+str(P[winner]))
                
                if P[winner] < isSignificant  and P[winner] > Maybe:
                    outfile.write(re.split("to",f[16:])[0]+","+re.split("sec",re.split("to",f[16:])[1])[0]+","+"Maybe"+","+classNames[winner]+","+str(P[winner]))
                            
                if P[winner] < Maybe:
                    outfile.write(re.split("to",f[16:])[0]+","+re.split("sec",re.split("to",f[16:])[1])[0]+","+"Insignificiant"+","+classNames[winner]+","+str(P[winner]))
                
                outfile.write("\n")
                
```
